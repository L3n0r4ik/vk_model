```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.metrics import ndcg_score
```

```python
#считываем данные
train_data = pd.read_csv("/kaggle/input/vkdeep/train_df.csv")
test_data = pd.read_csv("/kaggle/input/vkdeep/test_df.csv")
```

```python
#разделяем на X/y _train 
X_train = train_data.drop(['search_id', 'target'], axis=1)
y_train = train_data['target']
X_test = test_data.drop(['search_id', 'target'], axis=1)
y_test = test_data['target']
```

Попытка №1

```python
model = models.Sequential()
model.add(layers.Dense(128, activation='relu', input_shape=(79,)))  
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))  #Слой с сигмоидой
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=13, batch_size=32)
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Accuracy Test: {accuracy}')
```

```python
pred = model.predict(X_test)
```

```python
pred_classes = np.argmax(pred, axis=1)

ndcg = ndcg_score(np.expand_dims(y_test, axis=0), np.expand_dims(pred_classes, axis=0))
print("NDCG Score:", ndcg)
```
